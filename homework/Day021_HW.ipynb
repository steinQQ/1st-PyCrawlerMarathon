## 1. 取出今天所有的新聞

```python
from selenium import webdriver
from bs4 import BeautifulSoup
import os
import time
import re
import datetime

browser = webdriver.Chrome(os.getcwd() + '\chromedriver')

browser.get("https://www.ettoday.net/news/news-list.htm")
#每1秒往下滑，滑5次
import time
for i in range(5):
    time.sleep(1)
    browser.execute_script("window.scrollTo(0,10000);")
# 取得資料，丟到 BeautifulSoup 解析

html_soure = browser.page_source
browser = quit()
soup = BeautifulSoup(html_soure,"html5lib")
titles = soup.find('div',class_='part_list_2').find_all('h3')

x = datetime.datetime.now()#現在時間
y = "{}/{}/{}".format(x.year,x.month,x.day)


for title in titles :
    date = title.find('span')
    cls = title.find('em')
    tt = title.find('a')
    if  y in date.text :
        print(date.text,cls.text,tt.text)
```

## 2. 取出現在時間兩小時內的新聞

```python
from selenium import webdriver
from bs4 import BeautifulSoup
import os
import time
import re
import datetime
from datetime import timedelta

browser = webdriver.Chrome(os.getcwd() + '\chromedriver')
url = 'https://www.ettoday.net/news/news-list.htm'
browser.get(url)

for i in range(5):
    browser.execute_script('window.scrollTo(0,100000)')
    time.sleep(0.5)
    
html = browser.page_source
browser.quit()
soup = BeautifulSoup(html , 'html.parser')
titles=soup.find('div',class_='part_list_2').find_all('h3')

x = datetime.datetime.now() #現在時間
y = "{}/{}/{} {}".format(x.year,x.month,x.day,x.hour)
delta = timedelta(hours=1)
x1 = x - delta
y1 = "{}/{}/{} {}".format(x1.year,x1.month,x1.day,x1.hour)
x2 = x1 - delta
y2 = "{}/{}/{} {}".format(x2.year,x2.month,x2.day,x2.hour)

for title in titles :
    date = title.find('span')
    cls = title.find('em')
    tt = title.find('a')
    if  y in date.text or y1 in date.text or y2 in date.text:
        print(date.text,cls.text,tt.text)
```

## 3. 根據範例，取出三天前下午三點到五點的新聞

```python
from datetime import timedelta

browser = webdriver.Chrome(os.getcwd() + '\chromedriver')

x = datetime.datetime.now() #現在時間

url = "https://www.ettoday.net/news/news-list-{}-{}-{}-0.htm".format(x.year,x.month,x.day-3)   ###'https://www.ettoday.net/news/news-list.htm'
browser.get(url)

for i in range(20):
    browser.execute_script('window.scrollTo(0,100000)')
    time.sleep(0.5)
    
html = browser.page_source
browser.quit()
soup = BeautifulSoup(html , 'html.parser')
titles=soup.find('div',class_='part_list_2').find_all('h3')


y1 = "{}/{}/{} {}".format(x.year,x.month,x.day-3,15)
y2 = "{}/{}/{} {}".format(x.year,x.month,x.day-3,16)

for title in titles :
    date = title.find('span')
    cls = title.find('em')
    tt = title.find('a')
    if  y1 in date.text or y2 in date.text:
        print(date.text,cls.text,tt.text)
```
